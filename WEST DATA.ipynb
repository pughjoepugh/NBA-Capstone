```python
#https://www.basketball-reference.com/contracts/salary-cap-history.html

cap_history="""
Year	Salary Cap	Salary Cap (2019)
2016-17	$94,143,000	$98,194,497
2017-18	$99,093,000	$100,892,064
2018-19	$101,869,000	$101,869,000
2019-20	$109,140,000	$66777666
2020-21	$109,140,000	$64746464"""

####
# taking cap information, turning it into a list

# empty list
empty=[]

# clean up some characters
cap_replace=cap_history.replace(",",'_').replace('$',"").replace('	', '*').replace('\n', '*')

# break into individual words
cap_split=cap_replace.split('*')

# only keep desired words, happen to be sized 3 and up, adding them to the empty list 'empty'
for item in cap_split:
    if len(item)>3:
        empty.append(item)

# create a dictionary for salary cap numbers

cap = {empty[i]: float(empty[i + 1]) for i in range(3, len(empty), 3)}
```

```python
#player salaries

# formating the text salary information into a dataframe


# function for repeated text formating
def player_salary(tbl):
    
    #turn copied text/table into dictionary
    
    # list to store player information
    new = []
    
    # the same text cleaning as cap info. removing undesired characters and breaking into words
    
    REPLACED = tbl.replace(",",'_').replace('$',"").replace('\n', '*').replace('	','*').replace('Salary','s')
    BROKEN = REPLACED.split('*')
    for item in BROKEN:
        if item == "0":
            item = "0000"
        if len(item)>3:
            new.append(item)
    # dictionary for paired information
    
    res_dict = {new[i]: new[i + 1] for i in range(0, len(new)-3, 2)}
    
    #turn resulting dictionary into dataframe

    player_salary_df = pd.DataFrame.from_dict(res_dict, orient = 'index').reset_index().rename(columns = {'index':'Player',0:'Salary'})
    
    
    # create new columns from other text information
    
    # The Season
    player_salary_df['Season'] = new[-1]
    
    # The Team name
    player_salary_df['Team'] = new[-2]
    
    # The Win total for that season
    player_salary_df['Wins'] = new[-3]
    
    
    #clean up a little
    
    # adjusting index
    player_salary_df.index += 1
    
    # removing player names
    player_salary_df = player_salary_df.drop(columns = ['Player'])
    
    # changing data type for salaries
    player_salary_df[player_salary_df.columns[0]] = player_salary_df[player_salary_df.columns[0]].replace('_','', regex=True).astype(float)
    
    
    #create more columns of interest
    
    # Creating column for each player as a percent of total salary
    player_salary_df["% Total Salary"] = player_salary_df.Salary / player_salary_df.Salary.sum()*100
    
    player_salary_df["% Total Salary"] = player_salary_df["% Total Salary"].cumsum()
    
    player_salary_df[player_salary_df.columns[0]]=player_salary_df[player_salary_df.columns[0]].cumsum()
    
    
    # Creating column for each player as a percent of salary cap
    player_salary_df["% Salary Cap"]= player_salary_df.Salary / cap[new[-1]] *100
    
    
    #clean up, reshape
    
    # transpose so each column is a feature
    player_salary_df=player_salary_df.T
    
    # each team can only roster 15 players
    player_salary_df=player_salary_df.iloc[:,0:10]
    
    #reorganize
    
    team=player_salary_df.iloc[2:3]
    season=player_salary_df.iloc[1:2]
    test=player_salary_df.iloc[[0,3,4,5],:]
    test2=pd.concat([team,season,test])
    
    return test2
```
