```python
#player salaries

def player_salary(tbl):
    
    #turn copied table into dictionary
    
    new=[]
    REPLACED=tbl.replace(",",'_').replace('$',"").replace('\n', '*').replace('	','*').replace('Salary','s')
    BROKEN=REPLACED.split('*')
    for item in BROKEN:
        if item=="0":
            item="0000"
        if len(item)>3:
            new.append(item)
    res_dict = {new[i]: new[i + 1] for i in range(0, len(new)-3, 2)}
    
    #turn resulting dictionary into dataframe

    player_salary_df = pd.DataFrame.from_dict(res_dict, orient='index').reset_index().rename(columns={'index':'Player',0:'Salary'})
    
    #add columns for...
    
    player_salary_df['Season']= new[-1]
    player_salary_df['Team']= new[-2]
    player_salary_df['Wins']=new[-3]
    
    #clean up a little
    
    player_salary_df.index+=1
    player_salary_df=player_salary_df.drop(columns=['Player'])
    player_salary_df[player_salary_df.columns[0]]=player_salary_df[player_salary_df.columns[0]].replace('_','', regex=True).astype(float)
    
    #create more columns of interest
    
    player_salary_df["% Total Salary"]= player_salary_df.Salary / player_salary_df.Salary.sum()*100
    player_salary_df["% Total Salary"]=player_salary_df["% Total Salary"].cumsum()
    player_salary_df[player_salary_df.columns[0]]=player_salary_df[player_salary_df.columns[0]].cumsum()
    player_salary_df["% Salary Cap"]= player_salary_df.Salary / cap[new[-1]] *100
    
    #clean up, reshape
    
    player_salary_df=player_salary_df.T
    player_salary_df=player_salary_df.iloc[:,0:10]
    
    #reorganize
    
    team=player_salary_df.iloc[2:3]
    season=player_salary_df.iloc[1:2]
    test=player_salary_df.iloc[[0,3,4,5],:]
    test2=pd.concat([team,season,test])
    return test2
```
